```python
#creando mi propia lista, tupla, conjunto, diccionario

# mi creacion de diccionario se basara en un caso practico que sucede a menudo en la estructuracion de presupuesto

dict = {
    "VARI3/4"  : "varilla de acero de tres cuartos" ,
    "LADCOMUN" : "ladrillo de arcilla rectangular" ,
    "EQPSOLD"  : "equipo de soldadura" ,
    "ACEREFOR" : "acero reforzado" ,
    "CONCRE3000PSI" : "concreto de 3000 PSI"
    }
print(dict)

list_materiales_de_construccion = ["LADCOMUN" , "VARI3/4" , "EQPSOLD" , "ACEREFOR" , "CONCRE3000PSI" , "EQPSOLD"]
print(f"lista{list_materiales_de_construccion}")

tuple_materiales_de_construccion = ("LADCOMUN" , "VARI3/4" , "EQPSOLD" , "ACEREFOR" , "CONCRE3000PSI" , "EQPSOLD")
print(f"tuple{tuple_materiales_de_construccion}")

set_materiales_de_construccion = {"LADCOMUN" , "VARI3/4" , "EQPSOLD" , "ACEREFOR" , "CONCRE3000PSI" , "EQPSOLD"}
print(set_materiales_de_construccion)
```

creando mi for

```python
for material in list_materiales_de_construccion:
    print(material)

for clave, significado in dict.items():
    print(clave, significado)
```
