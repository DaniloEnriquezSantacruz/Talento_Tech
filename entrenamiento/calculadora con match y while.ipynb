```python
import math

def mostrar_titulo():
  print("CALCULADORA COMPLEJA")

def menu():
  print("\nElija una opcion:\n")
  print("1 => Suma")
  print("2 => Resta")
  print("3 => Multiplicacion")
  print("4 => Division")
  print("5 => Potencia")
  print("6 => Raiz Cuadrada")
  print("7 => logaritmo")
  print("8 => seno")
  print("9 => coseno")
  print("10 => tangente")
  print("0 => Salir\n")

def opcion():
  return int(input("Ingrese una opcion: "))

def pedir_dos_datos():
  num1 = int(input("Ingrese el primer numero: "))
  num2 = int(input("Ingrese el segundo numero: "))
  return num1, num2

def pedir_un_dato():
  num1 = int(input("Ingrese un numero: "))
  return num1

def suma(num1, num2):
  return num1 + num2

def resta(num1, num2):
  return num1 - num2

def multiplicacion(num1, num2):
  return num1 * num2

def division(num1, num2):
  return num1 / num2

def potencia(num1, num2):
  return math.pow(num1, num2)

def raiz_cuadrada(num1, num2):
  return math.pow(num1, 1/num2)

def logaritmo(num1):
  return math.log10(num1)

def seno(num1):
  return math.sin(num1)

def coseno(num1):
  return math.cos(num1)

def tangente(num1):
  return math.tan(num1)

def procesar_operaciones(opcion):

  match opcion:

    case 1:
      num1, num2 = pedir_dos_datos()
      print(f"La suma de {num1} y {num2} es: {suma(num1, num2)}")
    case 2:
      num1, num2 = pedir_dos_datos()
      print(f"La resta de {num1} y {num2} es: {resta(num1, num2)}")
    case 3:
      num1, num2 = pedir_dos_datos()
      print(f"La multiplicacion de {num1} y {num2} es: {multiplicacion(num1, num2)}")
    case 4:
      num1, num2 = pedir_dos_datos()
      if num2 == 0:
        print("No se puede dividir por cero")
      else:
        print(f"La division de {num1} y {num2} es: {division(num1, num2)}")
    case 5:
      num1, num2 = pedir_dos_datos()
      print(f"La potencia de {num1} elevado a {num2} es: {potencia(num1, num2)}")
    case 6:
      num1, num2 = pedir_dos_datos()
      print(f"La raiz cuadrada de {num1} y el indice como el numero {num2} es: {raiz_cuadrada(num1, num2)}")
    case 7:
      num1 = pedir_un_dato()
      if num1 <= 0:
        print("No se puede calcular el logaritmo de un numero negativo y tampoco del numero cero")
      else:
        print(f"El logaritmo de {num1} en base 10 es: {logaritmo(num1)}")
    case 8:
      num1 = pedir_un_dato()
      print(f"El seno de {num1} en radianes es: {seno(num1)}")
    case 9:
      num1 = pedir_un_dato()
      print(f"El coseno de {num1} en radianes es: {coseno(num1)}")
    case 10:
      num1 = pedir_un_dato()
      print(f"La tangente de {num1} en radianes es: {tangente(num1)}")

def salir():
  print("Gracias por usar la calculadora compleja")

def ejecucion_codigo():
  mostrar_titulo()
  menu()
  while True:
    try:
      opcion_elegida = opcion()
      if 0 <= opcion_elegida <= 10:
        procesar_operaciones(opcion_elegida)
        if opcion_elegida == 0:
          salir()
          break
      else:
        print("Opcion no valida")
      continuar = input("¿Desea realizar otra operacion? (s/n): ")
      if continuar.lower() != "s":
        salir()
        break
    except ValueError:
      print("Por favor, ingresa un número válido.")

ejecucion_codigo()
```
