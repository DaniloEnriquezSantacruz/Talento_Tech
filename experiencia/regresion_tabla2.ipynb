```python
import pandas as pd

y_nor = [78.5, 74.3, 104.3, 87.6, 95.9, 109.2, 102.7, 72.5, 93.1, 115.9, 83.8, 113.3, 109.4]

x1_list = [7, 1, 11, 11, 7, 11, 3, 1, 2, 21, 1, 11, 10]
x2_list = [26, 29, 56, 31, 52, 55, 71, 31, 54, 47, 40, 66, 68]
x3_list = [6, 15, 8, 8, 6, 9, 17, 22, 18, 4, 23, 9 ,8]
x4_list = [60, 52, 20, 47, 33, 22, 6, 44, 22, 26, 34, 12, 12]

data = {
    'x1': x1_list,
    'x2': x2_list,
    'x3': x3_list,
    'x4': x4_list
}

df = pd.DataFrame(data)

# Crea una nueva columna llamada 'columna_de_unos' con el valor 1 en todas las filas
df['y_unos'] = 1

# Crea una lista con el nuevo orden de las columnas
# Pon 'columna_de_unos' al principio, seguido de las columnas existentes
columnas_reordenadas = ['y_unos'] + [col for col in df.columns if col != 'y_unos']

# Reordena el DataFrame usando la lista de columnas reordenadas
X = df[columnas_reordenadas]

# Muestra el DataFrame para ver la nueva columna en la primera posici√≥n
X
```

```python
# Traspone el DataFrame
Xt = X.T

# Muestra el DataFrame traspuesto
Xt
```

```python
# Assuming X and Xt are already defined as in your previous code

# Realiza la multiplicaci√≥n de matrices usando el m√©todo .dot() en el orden correcto
res_mul = Xt.dot(X)

# Muestra el resultado de la multiplicaci√≥n
res_mul
```

```python
import numpy as np
inv = np.linalg.inv(res_mul)
inv = pd.DataFrame(inv)
inv
```
