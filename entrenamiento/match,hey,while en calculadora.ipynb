```python
def saludar(hora):
  if 6 <= hora < 12:
    print("buenos dias")
  elif 12 <= hora <= 18:
    print("buenas tardes")
  elif 19 <= hora <= 24 or 0 <= hora < 6:
    print("buenas noches")
  else:
    print("ingrese una hora valida")


hora = int(input("ingrese la hora actual: "))


saludar(hora)
```
