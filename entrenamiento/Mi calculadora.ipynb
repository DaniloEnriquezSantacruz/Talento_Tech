##Calculadora

```python
def calculadora():
    operacion = input("Selecciona una operación (+, -, *, /): ")
    num1 = float(input("Introduce el primer número: "))
    num2 = float(input("Introduce el segundo número: "))

    if operacion == "+":
        resultado = num1 + num2
    elif operacion == "-":
        resultado = num1 - num2
    elif operacion == "*":
        resultado = num1 * num2
    elif operacion == "/":
        if num2 == 0:
            return "Error: No se puede dividir entre cero."
        else:
            resultado = num1 / num2
    else:
        return "Operación inválida."

    return f"El resultado es: {resultado}"

print(calculadora())
```

#Calculadora mas Avanzada

```python
import math

def calculadora():
    operacion = input("Selecciona una operación (+, -, *, /, ^,num2√num1): ")
    num1 = float(input("Introduce el primer número: "))
    num2 = float(input("Introduce el segundo número: "))

    if operacion == "+":
        resultado = num1 + num2
    elif operacion == "-":
        resultado = num1 - num2
    elif operacion == "*":
        resultado = num1 * num2
    elif operacion == "/":
        if num2 == 0:
            return "Error: No se puede dividir entre cero."
        else:
            resultado = num1 / num2
    elif operacion == "potenciacion":
        print("elegiste potenciacion")
        num1 = float(input("Introduce el primer número: "))
        num2 = float(input("Introduce el segundo número: "))
        resultado = num1 ** num2
    elif operacion == "raiz":
        resultado = math.pow(num1, num2)
    elif operacion == "sin":
        resultado = math.sin(num1)
    elif operacion == "cos":
        resultado = math.cos(num1)
    elif operacion == "tan":
        resultado = math.tan(num1)
    elif operacion == "log"
        resultado = math.log(num1)
        print("elegiste logaritmo")
        num1 = float(input("Introduce el primer número: "))
        num2 = float(input("Introduce el segundo número: "))
        resultado = math.log(num1, num2)
    else:
        return "Operación inválida."

    return f"El resultado es: {resultado}"

print(calculadora())
```
